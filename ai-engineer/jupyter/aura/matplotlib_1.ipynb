```python
import matplotlib.pyplot as plt
import random

x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]

y_beijing = [random.uniform(5,10) for i in x]

plt.figure(figsize=(20,8),dpi=80)

plt.plot(x,y_shanghai, color='r',label="Shanghai")
plt.plot(x,y_beijing, color='g',label="Beijing")

## 标签
plt.xlabel('Time Trend')
plt.ylabel('Tempature Trend')
plt.title('The city tempature change')

## 修改刻度
x_label = ["11:{}".format(i) for i in x]
plt.xticks(x[::2],x_label[::2])
plt.yticks(range(0,40,5))

plt.grid(True, linestyle='--',alpha=0.5)

plt.legend()


plt.show()
```

```python
import matplotlib.pyplot as plt
import random



x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(5,10) for i in x]


figure,axes = plt.subplots(nrows=1,ncols=2)
plt.figure(figsize=(50,50),dpi=100)

axes[0].plot(x,y_shanghai, color='r',label="Shanghai")
axes[1].plot(x,y_beijing, color='g',label="Beijing")

axes[0].legend()
axes[1].legend()

## 标签
axes[0].set_xlabel('Time Trend')
axes[0].set_ylabel('Tempature Trend')
axes[1].set_xlabel('Time Trend')
axes[1].set_ylabel('Tempature Trend')

axes[0].set_title('The city tempature change')
axes[1].set_title('The city tempature change')

## 修改刻度
x_label = ["11:{}".format(i) for i in x]
axes[0].xticks(x[::2],x_label[::2])
axes[0].yticks(range(0,40,5))
axes[1].xticks(x[::2],x_label[::2])
axes[1].yticks(range(0,40,5))

axes[0].grid(True, linestyle='--',alpha=0.5)
axes[1].grid(True, linestyle='--',alpha=0.5)




axes[0].show()
axes[1].show()



```

```python
import matplotlib.pyplot as plt

x = ['shanghai','beijing','hangzhou','guangzhou']
y = [100,89,68,59]


plt.figure()


x_ticks = range(len(x))
plt.xticks(x_ticks,x)         
plt.bar(x_ticks,y,color=['r','g','y','b'])
plt.title('Title')
plt.grid(linestyle='--',alpha=0.5)

plt.show()


```

```python
import matplotlib.pyplot as plt

x = ['shanghai','beijing','hangzhou','guangzhou']
y_1 = [100,89,68,89]
y_2 = [120,99,98,69]


plt.figure()

x_ticks = range(len(x))
plt.xticks([i+0.11 for i in x_ticks],x)     
plt.bar(x_ticks,y_1,width=0.2)
plt.bar([i+0.22 for i in x_ticks],y_2,width=0.2)

plt.title('Title')
plt.grid(linestyle='--',alpha=0.5 )

plt.show()


```
